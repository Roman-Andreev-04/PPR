```python
# Установка pytrends, если еще не установлен
!pip install pytrends
import pandas as pd
from pytrends.request import TrendReq
import time # Импортируем модуль time, если понадобится для задержек

# Инициализация объекта TrendReq
pytrends = TrendReq(hl='en-US', tz=360)

# --- Формы для параметров запроса ---
# Ключевые слова для поиска (можно ввести несколько, разделив запятой)
keywords_for_search = "nishes" # @param {type:"string", title:"Ключевые слова"} # Значение по умолчанию
kw_list = [kw.strip() for kw in keywords_for_search.split(',') if kw.strip()] # Разделяем ввод на список ключевых слов

# Категория поиска (числовой ID). 0 = Все категории. См. пояснение ниже.
category_id = "0" # @param {type:"string", title:"Категория поиска (ID)"} # ID категории (например, '27' для 'Автомобили')

# Временной интервал (например, 'today 5-y' для последних 5 лет, '2020-01-01 2020-12-31' для конкретного периода)
time_range = "today 5-y" # @param {type:"string", title:"Временной интервал"} # Временной интервал

# Географический регион (например, 'US' для США, 'RU' для России, или оставить пустым для всего мира)
geographical_region = "" # @param {type:"string", title:"Географический регион"} # Географический регион

# Проверка, что список ключевых слов не пуст
if not kw_list:
    print("Пожалуйста, введите хотя бы одно ключевое слово в форму.")
else:
    try:
        # Построение полезной нагрузки для запроса
        # category_id преобразуется в int
        pytrends.build_payload(kw_list, cat=int(category_id), timeframe=time_range, geo=geographical_region, gprop='')

        # Получение связанных запросов
        related_queries = pytrends.related_queries()

        # Извлечение и вывод 10 самых популярных запросов
        top_queries_df = None
        for kw in kw_list:
            if kw in related_queries and 'top' in related_queries[kw] and related_queries[kw]['top'] is not None:
                top_queries_df = related_queries[kw]['top']
                break

        if top_queries_df is not None and not top_queries_df.empty:
            # Формируем строку для вывода ключевых слов
            keywords_display = "'" + "' / '".join(kw_list) + "'"
            print(f"10 самых популярных запросов по теме {keywords_display}:")
            for index, row in top_queries_df.head(10).iterrows():
                print(f"- {row['query']}")
        else:
            print("Не удалось найти данные по связанным запросам. Попробуйте изменить ключевые слова или параметры запроса.")

    except Exception as e:
        if "429" in str(e):
            print(f"Произошла ошибка: Google вернул код 429 (Too Many Requests).")
            print("Это означает, что вы отправили слишком много запросов за короткий промежуток времени.")
            print("Пожалуйста, подождите несколько минут (обычно 5-10) и попробуйте запустить ячейку снова.")
        else:
            print(f"Произошла ошибка: {e}")
            print("Пожалуйста, убедитесь, что у вас есть подключение к интернету и библиотека pytrends установлена.")
```
